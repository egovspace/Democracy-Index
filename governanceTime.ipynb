### Choropleth on World Governance

#### Source: Wilhelmsen_Vanhanen_Polyarchy
#### https://www.prio.org/Data/Governance/



```python
import pandas as pd
import plotly
import plotly.express as px
import plotly.offline as offline

```

```python
df = pd.read_excel('govData_v2.xlsx')
```

```python
fig = px.choropleth(df, locations="Alpha-3", color="ID", hover_name="Country", animation_frame="Year",
             color_continuous_scale=px.colors.sequential.algae, width=1200, height=600)
# fig.write_html('Democracy Index.html')
fig.show()
```
